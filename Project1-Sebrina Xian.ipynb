## Question 1 - The "Dream Score"  (5 marks)

Write a function dream_score(matrix) that takes in a list of lists, and returns the maximum possible score as an integer. You may assume the following:
<ul>
    <li>The turtle can "teleport" to any cell.</li>
    <li>Each cell can only be visited "once".</li>
    <li>You do not need to visit each cell (such as traps).</li>
    <li>There may be an item or trap on the goal.</li>
</ul>
<strong>Please note that the above assumptions 1 - 3 are solely relevant to this particular task only. Assumption 4 will be present throughout all the questions.</strong>

Eve will always travel from the cave entrance to the goal, so items or traps on the goal (assumption 4) must still be taken into account.

The parameters to this function are as follows:
<ul>
  <li><strong>matrix</strong> is a list of lists, where the "outer lists" denote rows and "inner lists" denote columns. Each cell is given an integer value denoting the "score" in the cell.</li>
</ul>
Input Assumptions:
<ul>
  <li>
    You can assume that the input arguments are syntactically correct given the definitions, and will always be non-empty (i.e a minimum grid size of 1×1).</li>
</ul>

Here are example calls to your function:

```python
>>> 3150
dream_score([[0, 150, 150, 150], [150, 150, 150, 150], [150, 150, 150, 150], [150, 150, 150, 150]])

dream_score([[0, 0, 0], [0, -1000, 0], [-150, 0, 0]])

```

````python
def dream_score(matrix):
    """
    This is a function that returns the max possible score obtainable in any N by N matrix for Eve
    """
    # Initialise i, j index, and points variable. so matrix[j][i] means Eva always starts from the cave entrance.
    # "outer lists"(i) denote rows and "inner lists"(j) denote columns. 
    # points variable means the scores Eve get by visit each ancient treasure
    i = 0
    j = 0
    points = 0
   
    # Check each location in the matrix, to see if the value on that location is negative value or zero, 
    # if yes, without assign the value to points variable, continue to check the next.
    # if value is positive, count into points
    for j in range (0, len(matrix)):
        for i in range (0, len(matrix)): 
            if matrix[j][i] <= 0:
                continue
            else:
                points = matrix[j][i] + points
    
    # Check the end location (the goal), if the value on the goal is negative, add to the points make it as the final_score.
    # if the value is positive, which means in above for loop already added once this value. final_score is just current points.
    if matrix[j][i] <= 0:
        goal = matrix[j][i]
        final_score = points + goal
    else: 
        final_score = points
    
    return final_score
````

## Question 2 - Valid Path  (5 marks)
Write a function <strong>valid_path(n, path)</strong> that takes in the size of the grid and a travel path, and returns <strong>True</strong> if the path is valid and <strong>False</strong> if the path is invalid.

A path is only valid when Eve ends at the goal coordinate (N−1,N−1) and the path conists of only legal moves between coordinates. Remember that coordinates must be inside the grid and legal moves are 1 unit: Right (East), Down (South), Diagonal-Right-Down (South-East).

````python
def valid_path(N, path):
    """
    This is a function used to check whether a path is a valid path for N by N matrix. 
    A valid path is only valid when Eve ends at the goal coordinate (N-1, N-1) 
    and the path consists of only legal moves between coordinates.
    """
    i = 0
    # Check the length of input path fisrt, if length less than 2, means Eve cannot end at the goal since it's always starts at entrance (0, 0), hence is False path.
    # If path length longer than 2, then continue 
    if len(path) < 2:
        return False
    else:
        valid_path = []
        # Check if the path start(entrance) is from (0, 0) and ends at the goal coordinate (N-1, N-1), if not, return False
        goal = path.pop()  
        entrance = path.pop(0) 
        if (goal == (N-1,N-1)) and (entrance == (0,0)):    # If the path starts at (0, 0) and ends at (N-1, N-1), then continue
            path = [entrance] + path + [goal]
            current = path.pop(0)                          # Assign path[0] to current
            for i in range(len(path)):  
                (x, y) = current
                if x > N-1 or y > N-1:                     # If current location is in exceed Matrix, return False
                    return False
                else:
                    validmove_list = [(x+1, y),(x, y+1),(x+1, y+1)]  # Based on the current location, generates three valid possible(East, South, South-east) coordinates
                    next = path.pop(0)                               # Get next position from path
                    
                    # Check if next position not the goal and in one of the valid moves in validmove_list list variable.
                    if (next != goal):                               # If next is not the goal and next is in valid move list, add current position to valid_path, update current location to next
                        if next in validmove_list:  
                            valid_path = valid_path + [current]
                            current = next
                        else:                                        # If next is not the goal but next is not in valid move list, return False 
                            return False
                    elif (next == goal) and (len(path) == 0):        # Else if next is the same as the goal and length of path is zero, then means finished all path check, and it is a valid path. 
                        return True
                    else:                                            # Else next is the same as the goal but there is still coodinates after current location reach the end. So it is still an invalid path. Return False.
                        return False
        else:
            return False
    return valid_path
````

````python
def greedy_path(matrix):
    # we always start from (0, 0) and end on (N - 1, N - 1)
    global N
    N = len(matrix)
    #global goal
    #goal = (N-1, N-1)
    # initialise a list to keep track of the current path
    # we start from (0, 0) so we can initialise it like this     
    start = (0, 0)
    global curr_best_score, curr_bestscore_location, score, path
    path = [start]
    score = [0]
    final_score = 0
    curr_best_score = 0
    curr_bestscore_location = start
       
    for i in range(0, 2*N):
        curr_bestscore_location = curr_bestscore_location
        possible_moves(curr_bestscore_location)
            
        if len(valid_moves) == 3:
            
            # get three possible direction moves coordinates
            east_coordinate = valid_moves[0]
            south_coordinate = valid_moves[1]
            southeast_coordinate = valid_moves[2]
        
            # get values on three possible moves:
            east_value = matrix[valid_moves[0][1]][valid_moves[0][0]]
            south_value = matrix[valid_moves[1][1]][valid_moves[1][0]]
            southeast_value = matrix[valid_moves[2][1]][valid_moves[2][0]]
            
            if (east_coordinate[0] in range (0, N)) & (east_coordinate[1] in range (0, N)):
                curr_best_score = east_value
                curr_bestscore_location = east_coordinate
            if (south_coordinate[0] in range (0, N)) & (south_coordinate[1] in range (0, N)):
                if south_value > curr_best_score:
                    curr_best_score = south_value
                    curr_bestscore_location = south_coordinate
            if (southeast_coordinate[0] in range (0, N)) & (southeast_coordinate[1] in range (0, N)):
                if southeast_value > curr_best_score:
                    curr_best_score = southeast_value
                    curr_bestscore_location = southeast_coordinate
            score = score + [curr_best_score]
            path = path + [curr_bestscore_location]
            
        elif len(valid_moves) == 1:
            curr_bestscore_location = valid_moves[0]
            curr_best_score = matrix[valid_moves[0][1]][valid_moves[0][0]]
            score = score + [curr_best_score]
            path = path + [curr_bestscore_location]
    
    for i in range(0, len(score)):
        final_score = final_score + score[i]
                
    return path, final_score

def possible_moves(current_coordination):
    """
    Generate valid moves in order of preference
    """
    global valid_moves
    valid_moves = []
    # generate a list of potential moves   
    (x, y) = current_coordination
    validmove_list = [(x+1, y),(x, y+1),(x+1, y+1)]
    for members in validmove_list:
        if members <= (N-1, N-1):
            valid_moves = valid_moves + [members]
    return valid_moves
````
